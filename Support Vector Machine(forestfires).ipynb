```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('P:/python/forestfires.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data1 = data.copy()
```

```python
data1 = data1.drop(['month', 'day'], axis = 1)
data1.columns
```

```python
data1[data1.duplicated()]
```

```python
data1 = data1.drop(labels = [53,100,215,303,426,461,501,508], axis = 0).reset_index(drop = True)
```

```python
data1[data1.duplicated()]
```

```python
data1.iloc[50:55,:3]
```

```python
data1.size_category.value_counts()
```

```python
size_category = {'small' : 0, 'large' : 1}
data1['size_category'] = data1['size_category'].map(size_category)
```

```python
data1['size_category'].value_counts()
```

```python
data1.shape
```

```python
x = data1.iloc[:,:-1]
y = data1.iloc[:,-1]
print(x.head())
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 40)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

### Support Vector Machine model building

```python
model = SVC(kernel='rbf')
model.fit(x_train , y_train)
```

```python
preds = model.predict(x_test)
preds
```

```python
acc = accuracy_score(y_test, preds) * 100
print('accuracy is', acc)
```

```python
confusion_matrix(y_test, preds)
```

```python
print('accuracy is', ((95+3)/(95+30+3))*100)
```

```python
# poly
model_p = SVC(kernel='poly')
model_p.fit(x_train , y_train)
preds_p = model_p.predict(x_test)
acc_p = accuracy_score(y_test, preds_p) * 100
print('accuracy is', acc_p)
confusion_matrix(y_test, preds_p)
```

### Model building using Gamma and C 

```python
clf = SVC()
param_grid = [{'kernel':['rbf','linear', 'poly'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf, param_grid, cv=10)
gsv.fit(x_train, y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

### Model building using 'Gamma = 50', 'c = 15' and 'kernel = linear'

```python
model_gck = SVC(kernel = 'linear', gamma = 50, C = 15)
model_gck.fit(x_train , y_train)
```

```python
preds_gck = model_gck.predict(x_test)
acc_gck = accuracy_score(y_test, preds_gck) * 100
print('accuracy is', acc_gck)
confusion_matrix(y_test, preds_gck)
```
